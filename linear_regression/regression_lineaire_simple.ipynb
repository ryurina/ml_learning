```python
data = pd.read_csv('data_rls.csv')
data.head()
```

```python
from pycaret.regression import *
```

```python
s = setup(data,
          target="Weight",
          log_experiment=True,
          experiment_name="weight_prediction",
          log_plots=True)
```

```python
best = compare_models(sort="R2")
```

```python
print(best)
```

```python
evaluate_model(best)
```

```python
final_best = finalize_model(best)
final_best
```

```python
save_model(final_best, "Weight_Prediction")
```
